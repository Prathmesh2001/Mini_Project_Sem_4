To check for false data in dataset and correcting it

```python
import pandas as pd
import numpy as np
import pickle

df = pd.read_csv(r"E:\SE\Sem 4\Mini-Project\Datasets\car_details\car_v2_copy.csv")
df.head()
```

```python
min_mean0 = df['min_cost_price'].mean()
max_mean0 = df['max_cost_price'].mean()
print(min_mean0)
print(max_mean0)
```

```python
#Check the correlation of all columns with selling price
corr_matrix = df.corr()
print(corr_matrix["selling_price"].sort_values(ascending=False))
```

```python
# mark false for min_cost_price > max_cost_price 
# mark true for NaN and equal values too
comp = np.where((df['min_cost_price'] <= df['max_cost_price']) | (df['min_cost_price'].isnull()),True,False)
i=False
if i in comp:
    print("False Values")
else:
    print("No False Values")
```

```python
df['error'] = comp
error = df.loc[df['error']==False]
error
```

```python
# to know which car brands have how many false values
error['brand'].value_counts()
```

```python
#check if any car in error df has min_cost_price less than 1Cr
error.loc[error['min_cost_price'] < 10000000]
```

```python
#from above we can safely assume that we have to divide the min_cost_price for error by 100 (e.g. 43Cr -> 43L)
df.loc[df['error']==False, 'min_cost_price'] = (df['min_cost_price']/100)
df.loc[df['error']==False]
```

```python
df.drop(['error'],axis='columns')

#Once we correct the false values, we confirm that there are no such values remaining

comp = np.where((df['min_cost_price'] <= df['max_cost_price']) | (df['min_cost_price'].isnull()),True,False)
if i==False in comp:
    print(i)
else:
    print("No false values")
    
df.drop(['error'],axis='columns', inplace=True)
```

```python
min_mean0 = df['min_cost_price'].mean()
max_mean0 = df['max_cost_price'].mean()
print(min_mean0)
print(max_mean0)
```
