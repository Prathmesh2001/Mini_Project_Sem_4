# Analysis of brands

```python
brand_list = df['brand'].unique()
brands = pd.Series(brand_list)
print("\nTotal number of brands: ",brands.shape)
df['brand'].value_counts()
```

```python
# check if mean was found to be nan (if no value for a brand is given)
# x1 = df.loc[df['max_cost_price']]
x1_list = list(df['brand'].unique())

for x1s in x1_list:
    print(x1s, len(df.loc[(df['brand']== x1s)]), len(df.loc[(df['brand']== x1s)].isnull()))
```

```python
#function to find mean for each brand
def find_mean(car_brand):
    brand_df = df.loc[(df['brand']== car_brand)]
    
    brand_max = brand_df['max_cost_price'].mean()
    brand_min = brand_df['min_cost_price'].mean()
    print(car_brand, len(brand_df), brand_df['max_cost_price'].isnull().sum())
    df.loc[(df['brand']==car_brand) & (df['max_cost_price'].isnull()), 'max_cost_price'] = brand_max
    df.loc[(df['brand']==car_brand) & (df['min_cost_price'].isnull()), 'min_cost_price'] = brand_min
    #print(car_brand,len(brand_df))
```

```python
# find_mean('Maruti')
for l in brand_list:
    find_mean(l)
```
